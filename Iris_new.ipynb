# Mean

```python
fun = sum(map(lambda x: x, iris['sepal_length']))
mean_sep_len = fun/len(iris['sepal_length'])
mean_sep_len
```

# Median

```python
median_sep_len = sorted(iris['sepal_length'])
len_sep = len(iris['sepal_length'])
median = median_sep_len[int((len_sep/2)+1):int((len_sep/2)+2)]
median
```

# Mode

```python
mode_sep_len = iris['sepal_length'].value_counts().head(1).to_list()
```

# Variance

```python
var =0
for i in iris['sepal_length']:
    var += (i - mean_sep_len)**2

var = var/(len_sep-1)
    
```

# Std Deviation


```python
import math as m
std = m.sqrt(var)
std
```
